# 06 — Web export (refresh the frontend)

**Run this whenever the data changes.** Re-run the analysis notebooks (01-05) so
`outputs/*.json` is up to date, then run this notebook. It regenerates the site's
data files from those outputs, so the frontend always reflects the current numbers
— no stale snapshot, nothing edited by hand.

Flow: `notebooks 01-05` → `outputs/*.json` → **this step** → `site/` data → refresh browser.


```python
# --- setup: make the repo modules importable (same convention as notebooks 01-05) ---
import os, sys
from pathlib import Path

REPO_ROOT = Path.cwd()
if not (REPO_ROOT / "config.py").exists() and (REPO_ROOT.parent / "config.py").exists():
    REPO_ROOT = REPO_ROOT.parent          # launched from inside notebooks/
sys.path.insert(0, str(REPO_ROOT))
os.chdir(REPO_ROOT)
print("repo root :", REPO_ROOT)
```

```python
# --- regenerate the frontend data from outputs/ ---
# This reads outputs/curb_intelligence.json, curb_offenders.json, curb_trends.json
# (+ curb_trends.csv) and rewrites the site's data files:
#     site/assets/curb-data.js     (window.CURB — every page reads this)
#     site/api/_hotspots.json      (rows the GenAI assistant is grounded on)
#
# The frontend folder must live in the repo as ./site  (change SITE if yours differs).
from src.webexport import export

SITE = "site"
assert os.path.isdir(SITE), f"'{SITE}/' not found. Put the frontend folder there, or change SITE."
export(outputs="outputs", site=SITE)
```
